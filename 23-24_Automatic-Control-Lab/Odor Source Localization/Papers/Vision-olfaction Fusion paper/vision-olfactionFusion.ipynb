```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.transforms as transforms
from matplotlib.lines import Line2D

def robotTrajectory(time, x, y, behavior, name):
    behavior_colors = {
        'In': 'blue',
        'Out': 'red',
        'Vision': 'green',
        'ObstacleAvoidance': 'orange'
    }
    
    plt.figure(figsize=(4, 6))
    plt.xlim(-2, 4)
    plt.ylim(-5, 5)

    for i in range(len(behavior) - 1):
        plt.plot([x[i], x[i + 1]], [y[i], y[i + 1]], color=behavior_colors[behavior[i]], linestyle='-')

    plt.plot(x[0], y[0], marker='*', markersize=10, color='blue')


    plt.title(name)
    plt.xlabel('Y')
    plt.ylabel('X')
    plt.grid(True)

    plt.plot(0, 0, 'bo')
    plt.plot(1.3, 3.5, 'ro')
    # circle = plt.Circle((1.3, 3.5), .85, color='black', fill=False)
    circle = plt.Circle((1.3, 3.5), .95, color='black', fill=False)
    plt.gca().add_patch(circle)


    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)

    legend_lines = [
        Line2D([0], [0], color='blue', linewidth=2, linestyle='-'),
        Line2D([0], [0], color='red', linewidth=2, linestyle='-'),
        Line2D([0], [0], color='green', linewidth=2, linestyle='-'),
        Line2D([0], [0], color='orange', linewidth=2, linestyle='-')
    ]

    # Define behavior labels for the legend
    behavior_labels = ['Olfaction', 'Crosswind', 'Vision', 'Obstacle']

    plt.legend(legend_lines, behavior_labels, loc='upper right', fontsize='small')
    square1 = patches.Rectangle((-0.7, -0.4), 0.3, 0.3, linewidth=1, edgecolor='orange', facecolor='orange')
    center1 = (1.5, -1)
    rotation1 = transforms.Affine2D().rotate_deg(-12).translate(*center1)
    square1.set_transform(rotation1 + plt.gca().transData)
    plt.gca().add_patch(square1)

    square2 = patches.Rectangle((-.7, 1.2), 0.3, 0.3, linewidth=1, edgecolor='orange', facecolor='orange')
    center2 = (1.65, 0)
    rotation2 = transforms.Affine2D().rotate_deg(-1).translate(*center2)
    square2.set_transform(rotation2 + plt.gca().transData)
    plt.gca().add_patch(square2)

    rectangle = patches.Rectangle((-.5, -4.2), 3.3, 8.2, linewidth=2, edgecolor='black', facecolor='none')
    center_rect = (-4.1 + 4.1, -1.65 + 1.65)
    rotation_rect = transforms.Affine2D().rotate_deg(-10).translate(*center_rect)
    rectangle.set_transform(rotation_rect + plt.gca().transData)
    plt.gca().add_patch(rectangle)

    plt.gca().invert_xaxis()
    plt.axis('equal')

    plt.show()

def graph(data, name):
    time = data.time
    x = data.y
    y = data.x
    behavior = data.behavior
    name = name
    return robotTrajectory(time, x, y, behavior, name)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.transforms as transforms
from matplotlib.lines import Line2D

def combinedTrajectory(time1, x1, y1, behavior1, time2, x2, y2, behavior2, time3, x3, y3, behavior3, time4, x4, y4, behavior4, time5, x5, y5, behavior5, name):
    behavior_colors = {
        'In': 'blue',
        'Out': 'red',
        'Vision': 'green',
        'ObstacleAvoidance': 'orange'
    }
    
    plt.figure(figsize=(4, 6))
    plt.xlim(-2, 4)
    plt.ylim(-5, 5)

    for i in range(len(behavior1) - 1):
        plt.plot([x1[i], x1[i + 1]], [y1[i], y1[i + 1]], color=behavior_colors[behavior1[i]], linestyle='-')

    for i in range(len(behavior2) - 1):
        plt.plot([x2[i], x2[i + 1]], [y2[i], y2[i + 1]], color=behavior_colors[behavior2[i]], linestyle='-')

    for i in range(len(behavior3) - 1):
        plt.plot([x3[i], x3[i + 1]], [y3[i], y3[i + 1]], color=behavior_colors[behavior3[i]], linestyle='-')

    for i in range(len(behavior4) - 1):
        plt.plot([x4[i], x4[i + 1]], [y4[i], y4[i + 1]], color=behavior_colors[behavior4[i]], linestyle='-')

    for i in range(len(behavior5) - 1):
        plt.plot([x5[i], x5[i + 1]], [y5[i], y5[i + 1]], color=behavior_colors[behavior5[i]], linestyle='-')

    plt.plot(x1[0], y1[0], marker='*', markersize=10, color='blue')
    plt.plot(x2[0], y2[0], marker='*', markersize=10, color='blue')
    plt.plot(x3[0], y3[0], marker='*', markersize=10, color='blue')
    plt.plot(x4[0], y4[0], marker='*', markersize=10, color='blue')
    plt.plot(x5[0], y5[0], marker='*', markersize=10, color='blue')

    plt.title(name)
    plt.xlabel('Y')
    plt.ylabel('X')
    plt.grid(True)

    plt.plot(0, 0, 'bo')
    plt.plot(1.3, 3.5, 'ro')
    # circle = plt.Circle((1.3, 3.5), .85, color='black', fill=False)
    circle = plt.Circle((1.3, 3.5), .95, color='black', fill=False)
    plt.gca().add_patch(circle)

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)

    legend_lines = [
        Line2D([0], [0], color='blue', linewidth=2, linestyle='-'),
        Line2D([0], [0], color='red', linewidth=2, linestyle='-'),
        Line2D([0], [0], color='green', linewidth=2, linestyle='-'),
        Line2D([0], [0], color='orange', linewidth=2, linestyle='-')
    ]

    # Define behavior labels for the legend
    behavior_labels = ['Olfaction', 'Crosswind', 'Vision', 'Obstacle']

    plt.legend(legend_lines, behavior_labels, loc='upper right', fontsize='small')
    square1 = patches.Rectangle((-0.7, -0.4), 0.3, 0.3, linewidth=1, edgecolor='orange', facecolor='orange')
    center1 = (1.5, -1)
    rotation1 = transforms.Affine2D().rotate_deg(-12).translate(*center1)
    square1.set_transform(rotation1 + plt.gca().transData)
    plt.gca().add_patch(square1)

    square2 = patches.Rectangle((-.7, 1.2), 0.3, 0.3, linewidth=1, edgecolor='orange', facecolor='orange')
    center2 = (1.65, 0)
    rotation2 = transforms.Affine2D().rotate_deg(-1).translate(*center2)
    square2.set_transform(rotation2 + plt.gca().transData)
    plt.gca().add_patch(square2)

    rectangle = patches.Rectangle((-.5, -4.2), 3.3, 8.2, linewidth=2, edgecolor='black', facecolor='none')
    center_rect = (-4.1 + 4.1, -1.65 + 1.65)
    rotation_rect = transforms.Affine2D().rotate_deg(-10).translate(*center_rect)
    rectangle.set_transform(rotation_rect + plt.gca().transData)
    plt.gca().add_patch(rectangle)

    plt.gca().invert_xaxis()
    plt.axis('equal')

    plt.show()


def combinedGraph(data1, data2, data3, data4, data5, name):
    time1 = data1.time
    x1 = data1.y
    y1 = data1.x
    behavior1 = data1.behavior

    time2 = data2.time
    x2 = data2.y
    y2 = data2.x
    behavior2 = data2.behavior

    time3 = data3.time
    x3 = data3.y
    y3 = data3.x
    behavior3 = data3.behavior

    time4 = data4.time
    x4 = data4.y
    y4 = data4.x
    behavior4 = data4.behavior

    time5 = data5.time
    x5 = data5.y
    y5 = data5.x
    behavior5 = data5.behavior

    name = name
    return combinedTrajectory(time1, x1, y1, behavior1, time2, x2, y2, behavior2, time3, x3, y3, behavior3, time4, x4, y4, behavior4, time5, x5, y5, behavior5, name)
```
